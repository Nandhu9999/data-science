```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
df_=pd.read_csv('/content/supermarket_sales.csv')
df_
```

```python
df_.columns
```

```python
df=pd.DataFrame({'Branch':df_['Branch'],'Gender':df_['Gender'],'Product':df_['Product line'],'Total':df_['Total'],'Paymnet':df_['Payment'],})
df['Customer type']=df_['Customer type']
df
```

```python
df.groupby('Branch').mean()
```

```python
# H0: Branch A makes an averge total of 312 Dollars per day.
# Ha: Branch A makes more than 312 Dollars per day.
```

```python
sample=df.loc[df['Branch']=='A']
samplea=sample.sample(165)
samplea
```

```python
samplea.Total.mean()
```

```python
# standard deviation of the population
sd=np.std(samplea['Total'])
sd
```

```python
samplea.std()
```

```python
# Z TEST
xbar=samplea.Total.mean()
mu=312.3
n=200 #sample size
alpha=0.05
```

```python
xbar
```

```python
zvalue=(xbar-mu)/(sd/np.sqrt(n))
zvalue
```

```python
z_critical = st.norm.ppf(alpha)  #One tail Test
z_critical
```

```python
if (zvalue > z_critical and abs(zvalue) < abs(z_critical)):
  print("Cannot reject H0")
else:
  print("Reject H0")
```

```python
p_val = (1 - st.norm.cdf(abs(zvalue)))
p_val 
```

```python
#comparing with alpha value
if p_val<alpha:
  print("Reject H0")
else:
  print("Cannot reject H0")
```

```python
# T TEST
tsamplea=sample.sample(25)
tsamplea
```

```python
t_sd=np.std(tsamplea['Total'])
t_sd
```

```python
xbar=tsamplea.Total.mean()
print(xbar)
mu=312.3
n=50 #sample size
alpha=0.05
```

```python
n = 50
degrees_of_freedom = n-1
t = (xbar - mu)/(t_sd / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, degrees_of_freedom)
t_critical
```

```python
if (t > t_critical):
    print("Cannot be reject H0")
else:
    print("Reject H0")
```
